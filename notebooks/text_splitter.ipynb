```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

# 1. Load PDF (1 doc per page)
loader = PyPDFLoader("/Users/raghavendraggoudar/Desktop/AI-MATERIALS/RAG/rag_app/data/documents/attention.pdf")
docs = loader.load()  # List of Document objects
print(f"Loaded {len(docs)} pages")

# 2. Split into smaller chunks
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,      # Target chars per chunk
    chunk_overlap=200,    # Overlap for context
    length_function=len,
    add_start_index=True  # Track original position
)

chunks = text_splitter.split_documents(docs)
print(f"Split into {len(chunks)} chunks")
```

```python
lengths = [len(c.page_content) for c in chunks]
```
